```python
PS_Teachers = pd.read_csv("Teachers_export.csv", encoding='latin-1')
```

```python
PS_Teachers
```

```python
Grade2_Teachers = PS_Teachers.loc[PS_Teachers['Title'] == '2nd Grade Teacher']
```

```python
Grade2_Teachers.info()
```

```python
transform_schoolID = {
    100:'Maricopa',
    150: 'Casa Grande',
    200: 'Queen Creek',
    250: 'Avondale',
    300: 'Chandler',
    350: 'Northwest Tucson',
    400: 'Laveen',
    450: 'Gilbert',
    500: 'Surprise',
    550: 'North Chandler',
    600: 'Glendale',
    650: 'Peoria',
    700: 'East Mesa',
    750: 'Phoenix',
    800: 'Goodyear',
    850: 'North Phoenix',
    900: 'West Surprise',
    950: 'East Tucson',
    1000: 'Deer Valley',
    1050: 'Mesa',
    1100: 'Legacy Online Academy',
}

Grade2_Teachers['School/Building'] = Grade2_Teachers['School ID'].map(transform_schoolID)
```

```python
Grade2_Teachers['School/Building Code'] = 'LTS-AZ-' + Grade2_Teachers['School ID'].astype(str)
```

```python
Grade2_Teachers['Class'] = ''
Grade2_Teachers['Grade'] = '2'
Grade2_Teachers['Code'] = ''
```

```python
Grade2_Teachers
```

```python
Grade2_Teachers.info()
```

```python
df_StaffFile = Grade2_Teachers.filter(['School/Building', 'School/Building Code', 'Class', 'Grade', 'Code', 'First Name', 
                                       'Last Name', 'Email'])
```

```python
df_StaffFile['System *'] = 'Legacy Traditional School'
df_StaffFile['District/Area'] = 'Legacy Traditional Schools- AZ'
df_StaffFile['Middle Name'] = ''
df_StaffFile['Gender'] = ''
df_StaffFile['Password'] = ''
df_StaffFile['Middle Name'] = ''
df_StaffFile['Roles *'] = 'T'
df_StaffFile['Active User *'] = 'Y'
```

```python
df_StaffFile.rename(columns = {"First Name": "First Name *",'Last Name': 'Last Name *', 'Email': 'Email Address *'}, inplace=True)
```

```python
df_Staff = df_StaffFile[['System *','District/Area', 'School/Building', 'School/Building Code', 'Class', 'Grade', 'Code', 'First Name *',
             'Last Name *', 'Middle Name', 'Gender', 'Email Address *', 'Password', 'Roles *', 'Active User *']]
```

```python
df_Staff
```

```python
df_Staff.to_csv('Cogat_2ndGradeTeachers.csv', index=False, encoding='utf-8')
```
