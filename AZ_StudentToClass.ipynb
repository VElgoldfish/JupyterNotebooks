```python
AZ_Active_Classes = AZClass[((AZClass['CC.SECTIONID'] > 0) == True)]
```

```python
((AZ_Active_Classes['CC.SECTIONID'] < 0) == True).sum()
```

```python
AZ_Active_Classes['Primary Class ID'] = AZ_Active_Classes['CC.SECTIONID'].astype(str) + "-" + AZ_Active_Classes['Teacher Number'].astype(str)
```

```python
AZ_Active_Classes.info()
```

```python
df_HomeroomOnly = AZ_Active_Classes[(AZ_Active_Classes['Course Name'].str.contains('Homeroom', na=False, regex=True) == True)]
```

```python
transform_schoolID = {
    100:'Maricopa',
    150: 'Casa Grande',
    200: 'Queen Creek',
    250: 'Avondale',
    300: 'Chandler',
    350: 'Northwest Tucson',
    400: 'Laveen',
    450: 'Gilbert Elementary School',
    500: 'Surprise',
    550: 'North Chandler',
    600: 'Glendale',
    650: 'Peoria',
    700: 'East Mesa',
    750: 'Phoenix',
    800: 'Goodyear',
    850: 'North Phoenix',
    900: 'West Surprise',
    950: 'East Tucson',
    1000: 'Deer Valley',
    1050: 'Mesa',
    1100: 'Legacy Online Academy',
}

df_HomeroomOnly['Institution'] = df_HomeroomOnly['School Number'].map(transform_schoolID)
```

```python
df_HomeroomOnly.info()
```

```python
df_StudentToClass = df_HomeroomOnly.filter(['Institution', 'Primary Class ID', 'State Student number'])
```

```python
df_StudentToClass['Is Official Class'] = ''
```

```python
df_writecsv = df_StudentToClass.rename(columns={"School Name": "Institution", "State Student number": "Primary Student ID (State ID)"})
```

```python
df_writecsv
```

```python
df_writecsv.to_csv('Legacy_Traditional_Schools_Arizona_student_to_class.csv', index=False, encoding='utf-8')
```
