```python
NV_Student_Course = pd.read_csv("Student_Course_SQL.csv", encoding='latin-1')
```

```python
NV_Student_Course.info()
```

```python
NV_Class_Fields = NV_Student_Course.filter(['number','sectionID', 'staffnumber', 'coursename', 'periodstart', 'grade', 'teacherfullname'])
```

```python
NV_Class_Fields
```

```python
NV_Class_Fields.drop_duplicates()
```

```python
df_NV_Homeroom = NV_Class_Fields[NV_Class_Fields['coursename'].str.contains("Homeroom")]
```

```python
df_NV_Homeroom.drop_duplicates()
```

# Things to do: <br>
* 0K to grade 0 (check), coerce all to integers - done
* Create the "Class Name" field - done
* Create Primary Class ID - done
* Create "Institution" - done


```python
# Add institution name
# Change School ID to Institution
transform_schoolID = {
    39438: 'North Valley',
    39439: 'Cadence',
    39440: 'Southwest Las Vegas'
}

df_NV_Homeroom['Institution'] = df_NV_Homeroom['number'].map(transform_schoolID)
```

```python
df_NV_Homeroom['Primary Class ID'] = df_NV_Homeroom['sectionID'].astype(str) + '-' + df_NV_Homeroom['staffnumber'].astype(str)
```

```python
df_NV_Homeroom
```

```python
df_NV_Homeroom['Int_Grade'] = np.where(df_NV_Homeroom['grade'] == '0K', 0, df_NV_Homeroom['grade'])
```

```python
# Fix this by casting
df_NV_Homeroom.Int_Grade = pd.to_numeric(df_NV_Homeroom.Int_Grade, errors="coerce")
```

```python
df_NV_Homeroom.drop_duplicates()
```

```python
df_NV_Homeroom['Teacher_Last'] = df_NV_Homeroom['teacherfullname'].str.partition(', ')[0]
```

```python
df_NV_Homeroom['ClassNameNoCoursename'] = df_NV_Homeroom['coursename'].astype(str) + '-' + df_NV_Homeroom['Teacher_Last'].astype(str)
```

```python
df_NV_Homeroom['Class Name'] = df_NV_Homeroom['periodstart'].astype(str) + '-' + df_NV_Homeroom['coursename'].astype(str) + '-' + df_NV_Homeroom['Teacher_Last'].astype(str)
```

```python
df_NV_Homeroom.drop_duplicates()
```

```python
# Creating the dataframe next 
```

```python
df_ClassFile = df_NV_Homeroom.filter(['Institution', 'Primary Class ID', 'Class Name', 'Int_Grade'])
```

```python
df_ClassFile.drop_duplicates()
```

```python
df_ClassFile.rename(columns = {'Int_Grade': 'Grade'}, inplace=True)
```

```python
df_ClassCSV = df_ClassFile.drop_duplicates()
```

```python
df_ClassCSV.to_csv('Legacy_Traditional_Schools_Nevada_class.csv', index=False, encoding='utf-8')
```
