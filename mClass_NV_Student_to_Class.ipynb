<H4> Fields needed for Student to Class File are: <br>
    * Institution <br>
    * Primary Class ID <i> Will be sectionid + staffnumber </i> <br>
    * Primary Student ID (State ID) <br>
    * "Is Official Class" = ''

```python
NV_Student_Course = pd.read_csv("Student_Course_SQL.csv", encoding='latin-1')
```

```python
NV_Student_Course.info()
```

```python
NV_Class_Fields = NV_Student_Course.filter(['number','sectionID', 'staffnumber', 'stateID', 'periodstart', 'coursename'])
```

```python
df_NV_Homeroom = NV_Class_Fields[NV_Class_Fields['coursename'].str.contains("Homeroom")]
```

```python
df_NV_Homeroom 
```

```python
NV_Class_Fields.drop_duplicates()
```

```python
NV_Class_Fields.drop_duplicates()
```

```python
df_NV_Homeroom.drop_duplicates()
```

```python
# Add institution name
# Change School ID to Institution
transform_schoolID = {
    39438: 'North Valley',
    39439: 'Cadence',
    39440: 'Southwest Las Vegas'
}

df_NV_Homeroom['Institution'] = df_NV_Homeroom['number'].map(transform_schoolID)
```

```python
df_NV_Homeroom['Primary Class ID'] = df_NV_Homeroom['sectionID'].astype(str) + '-' + df_NV_Homeroom['staffnumber'].astype(str)
```

```python
df_NV_Homeroom
```

```python
# Creating the dataframe next 
```

```python
df_ClassFile = df_NV_Homeroom.filter(['Institution', 'Primary Class ID', 'stateID'])
```

```python
df_ClassFile.drop_duplicates()
```

```python
df_ClassFile.rename(columns = {'stateID': 'Primary Student ID (State ID)'}, inplace=True)
```

```python
df_ClassFile['Is Official Class']=''
```

```python
df_Student_to_ClassCSV = df_ClassFile.drop_duplicates()
```

```python
df_Student_to_ClassCSV
```

```python
df_Student_to_ClassCSV.to_csv('Legacy_Traditional_Schools_Nevada_student_to_class.csv', index=False, encoding='utf-8')
```
