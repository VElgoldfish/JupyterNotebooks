```python
# Drop all null course names, down to 1,519,599 rows
df_April_NotNull = df_April.dropna(subset=['course_name'])
```

```python
# Grammar, no filter: 69,933 rows Ã— 35 columns
df_coursename_grammar = df_April_NotNull.loc[df_April_NotNull['course_name'].str.contains("grammar", case = False)]
```

```python
df_coursename_grammar.info()
```

```python
# Science, no filter: 45,537 rows
df_coursename_science = df_April_NotNull.loc[df_April_NotNull['course_name'].str.contains("science", case = False)]
```

```python
df_coursename_science.info()
```

```python
# Reading, no filter: 69,860 rows
df_coursename_reading = df_April_NotNull.loc[df_April_NotNull['course_name'].str.contains("reading", case = False)]
```

```python
df_coursename_reading.info()
```

Necessary columns - to remove duplicates by day:
- role_name <br>

- course_name <br>
- date <br>


<H3> Sorting each subject by required fields

```python
# 7,618 rows
df_grammar_filter = df_coursename_grammar.filter(['role_name','schoology_user_id', 'user_building_name', 'course_name', 'date'])
df_grammar = df_grammar_filter.drop_duplicates()
```

```python
df_grammar
```

```python
# 11,554 rows
df_science_filter = df_coursename_science.filter(['role_name', 'schoology_user_id', 'user_building_name', 'course_name', 'date'])
df_science = df_science_filter.drop_duplicates()
```

```python
df_science
```

```python
# 8,050 rows
df_reading_filter = df_coursename_reading.filter(['role_name', 'schoology_user_id', 'user_building_name', 'course_name', 'date'])
df_reading = df_reading_filter.drop_duplicates()
```

```python
df_reading
```

```python
# pd.concat([t1, t2, t3, t4, t5])
df_April_ScienceReadingGrammar = pd.concat([df_reading, df_science, df_grammar])
```

```python
# 27,222 rows, as predicted
df_April_ScienceReadingGrammar
```

```python
df_April_ScienceReadingGrammar.to_csv('AprilScienceReadingGrammar.csv', encoding='utf-8', index = False)
```
