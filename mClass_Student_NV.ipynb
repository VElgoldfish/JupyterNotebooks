```python
# Get rid of 0K
NV_Homerooms['Int_Grade'] = np.where(NV_Homerooms['grade'] == '0K', 0, NV_Homerooms['grade'])
```

```python
# Fix this by casting
NV_Homerooms.Int_Grade = pd.to_numeric(NV_Homerooms.Int_Grade, errors="coerce")
```

```python
NV_Homerooms['Int_Grade'].value_counts()
```

```python
NV_Homerooms.info()
```

```python
NV_Homerooms['Int_Grade'] = NV_Homerooms['Int_Grade'].astype('int')
```

```python
NV_Homerooms['birthdate'] = pd.to_datetime(NV_Homerooms['birthdate'])
```

```python
NV_Homerooms['DOB'] = NV_Homerooms['birthdate'].dt.strftime('%m/%d/%Y')
```

```python
NV_Homerooms.info()
```

```python
NV_Homerooms
```

```python
NV_Homerooms['Suffix']=''
```

```python
NV_Homerooms['Email'] = 'nv' + NV_Homerooms['studentNumber'].astype(str) + '@legacystudent.org'
```

```python
NV_Homerooms
```

```python
NV_Homerooms['number'].value_counts().sort_index()
```

```python
transform_schoolID = {
    39438:'North Valley',
    39439: 'Cadence',
    39440: 'Southwest Las Vegas',
}

NV_Homerooms['Institution'] = NV_Homerooms['number'].map(transform_schoolID)
```
