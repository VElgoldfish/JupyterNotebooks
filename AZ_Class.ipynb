```python
import pandas as pd
import numpy as np
from io import StringIO
```

```python
AZClass = pd.read_csv("Course_Data_ELEM.csv", encoding='latin-1')
```

```python
AZ_Active_Classes = AZClass[((AZClass['CC.SECTIONID'] > 0) == True)]
```

```python
AZ_Active_Classes['Class Name'] = AZ_Active_Classes['Section Number'].astype(str) + "-" + AZ_Active_Classes['Course Name'].astype(str) + "-" + AZ_Active_Classes['Teacher Last'].astype(str)
```

```python
AZ_Active_Classes['Primary Class ID'] = AZ_Active_Classes['CC.SECTIONID'].astype(str) + "-" + AZ_Active_Classes['Teacher Number'].astype(str)
```

```python
AZ_Active_Classes
```

```python
transform_schoolID = {
    100:'Maricopa',
    150: 'Casa Grande',
    200: 'Queen Creek',
    250: 'Avondale',
    300: 'Chandler',
    350: 'Northwest Tucson',
    400: 'Laveen',
    450: 'Gilbert Elementary School',
    500: 'Surprise',
    550: 'North Chandler',
    600: 'Glendale',
    650: 'Peoria',
    700: 'East Mesa',
    750: 'Phoenix',
    800: 'Goodyear',
    850: 'North Phoenix',
    900: 'West Surprise',
    950: 'East Tucson',
    1000: 'Deer Valley',
    1050: 'Mesa',
    1100: 'Legacy Online Academy',
}

AZ_Active_Classes['Institution'] = AZ_Active_Classes['School Number'].map(transform_schoolID)
```

```python
AZ_Active_Classes['Institution']
```

```python
df_ClassFields = AZ_Active_Classes.filter(['Institution', 'Primary Class ID', 'Class Name','Grade Level'])
```

```python
df_ClassFields
```

<H4> Course Name should contain Homeroom only

```python
df_HomeroomClasses = df_ClassFields[df_ClassFields['Class Name'].str.contains("Homeroom")]
```

<H4> Make sure you have unique values

```python
df_UniqueClasses = df_HomeroomClasses.drop_duplicates()
```

```python
df_UniqueClasses
```

```python
df_writetocsv = df_UniqueClasses.rename(columns={"School Name": "Institution", "Course Name": "Class Name", "Grade Level": "Grade"})
```

```python
df_writetocsv
```

```python
df_writetocsv.to_csv('Legacy_Traditional_Schools_Arizona_class.csv', index=False, encoding='utf-8')
```
