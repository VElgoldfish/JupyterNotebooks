<H4> Rename columns per mClass template column headings

```python
AZ_FocusedColumns.rename(columns={'STUDENTS.SchoolID': 'SchoolID', 'STUDENTS.State_StudentNumber': 'State_Student_ID', 'STUDENTS.Student_Number': 'Student_Number', 'STUDENTS.Last_Name': 'Last_Name', 'STUDENTS.First_Name': 'First_Name', 'STUDENTS.Middle_Name': 'Middle_Name', 'STUDENTS.Grade_Level': 'Grade', 'STUDENTS.DOB': 'DOB', 'STUDENTS.Home_Room': 'Homeroom', 'STUDENTS.Gender': 'Gender'}, inplace=True)
```

```python
AZ_FocusedColumns
```

<H4> Check counts for SchoolID, and then rename

```python
AZ_FocusedColumns.SchoolID.value_counts().sort_index()
```

```python
transform_schoolID = {
    100:'Maricopa',
    150: 'Casa Grande',
    200: 'Queen Creek',
    250: 'Avondale',
    300: 'Chandler',
    350: 'Northwest Tucson',
    400: 'Laveen',
    450: 'Gilbert Elementary School',
    500: 'Surprise',
    550: 'North Chandler',
    600: 'Glendale',
    650: 'Peoria',
    700: 'East Mesa',
    750: 'Phoenix',
    800: 'Goodyear',
    850: 'North Phoenix',
    900: 'West Surprise',
    950: 'East Tucson',
    1000: 'Deer Valley',
    1050: 'Mesa',
    1100: 'Legacy Online Academy',
}

AZ_FocusedColumns['Institution'] = AZ_FocusedColumns['SchoolID'].map(transform_schoolID)
```

```python
AZ_FocusedColumns.Institution.value_counts().sort_index()
```

```python
# Check grades
AZGradesKthrough6 = AZ_FocusedColumns.loc[AZ_FocusedColumns['Grade'] <= 6]
```

```python
AZGradesKthrough6
```

```python
AZGradesKthrough6.reset_index(drop=True)
```

<H4> Need to change formatting for DOB MM-DD-YYYY

```python
df_Kthru6_DOB = AZGradesKthrough6.copy()
```

```python
df_Kthru6_DOB['DOB'] = pd.to_datetime(df_Kthru6_DOB.DOB)
```

```python
df_Kthru6_DOB['format_DOB'] = df_Kthru6_DOB['DOB'].dt.strftime('%m/%d/%Y')
```

```python
df_Kthru6_DOB
```

<H5> Add two columns: student email + suffix

```python
df_Kthru6_DOB['Student_Email'] = df_Kthru6_DOB['Student_Number'].astype(str) + '@legacystudent.org'
```

```python
df_Kthru6_DOB['Suffix'] = ''
```

```python
df_Kthru6_DOB
```
