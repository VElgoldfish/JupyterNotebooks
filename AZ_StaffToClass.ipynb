```python
AZClass = pd.read_csv("Course_Data_ELEM.csv", encoding='latin-1')
```

```python
PS_Active_Classes = AZClass[((AZClass['CC.SECTIONID'] > 0) == True)]
```

```python
PS_Active_Classes
```

<H4> Homeroom classes only

```python
df_HomeroomOnly = PS_Active_Classes[(PS_Active_Classes['Course Name'].str.contains('Homeroom', na=False, regex=True) == True)]
```

```python
df_HomeroomOnly
```

```python
df_HomeroomOnly['Primary Staff ID (State ID)'] = df_HomeroomOnly['School Number'].astype(str) + '-' + df_HomeroomOnly['Teacher Number'].astype(str)
```

```python
df_HomeroomOnly['Primary Class ID'] = df_HomeroomOnly['CC.SECTIONID'].astype(str) + "-" + df_HomeroomOnly['Teacher Number'].astype(str)
```

```python
transform_schoolID = {
    100:'Maricopa',
    150: 'Casa Grande',
    200: 'Queen Creek',
    250: 'Avondale',
    300: 'Chandler',
    350: 'Northwest Tucson',
    400: 'Laveen',
    450: 'Gilbert Elementary School',
    500: 'Surprise',
    550: 'North Chandler',
    600: 'Glendale',
    650: 'Peoria',
    700: 'East Mesa',
    750: 'Phoenix',
    800: 'Goodyear',
    850: 'North Phoenix',
    900: 'West Surprise',
    950: 'East Tucson',
    1000: 'Deer Valley',
    1050: 'Mesa',
    1100: 'Legacy Online Academy',
}

df_HomeroomOnly['Institution'] = df_HomeroomOnly['School Number'].map(transform_schoolID)
```

```python
df_HomeroomOnly
```

```python
# Rename appropriate headings: Institution, Primary Class ID, Primary Staff ID
```

```python
df_StaffToClass = df_HomeroomOnly.filter(['Institution', 'Primary Class ID', 'Primary Staff ID (State ID)'])
```

```python
df_StaffToClass
```

```python
df_StaffToClass['Is Primary Teacher']=''
```

```python
df_StaffToClass.drop_duplicates()
```

```python
df_writecsv = df_StaffToClass.rename(columns={"School Name": "Institution"})
```

```python
df_noduplicates = df_writecsv.drop_duplicates()
```
